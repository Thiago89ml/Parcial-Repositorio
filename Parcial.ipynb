```python
import pandas as pd
```

```python
df=pd.read_csv('train.csv')
```

```python
df.info()
```

```python
df.head(50)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['CryoSleep'] = df['CryoSleep'].map({'earth': 1,'republican':2,'mars':3}).fillna(1)
```

```python
df['VIP'] = df['VIP'].map({'False':1,'True':2}).fillna(0)
```

```python
df['FoodCourt'].fillna(df['FoodCourt'].median(),inplace=True)
```

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
```

```python
df['RoomService'].fillna(df['RoomService'].mode()[0],inplace = True)
df['ShoppingMall'].fillna(df['ShoppingMall'].mode()[0],inplace = True)
df['VRDeck'].fillna(df['VRDeck'].median(),inplace=True)
df.drop(columns='Destination',inplace=True)
df.drop(columns='Cabin', inplace=True)
df.drop(columns='PassengerId',inplace =True)


```

```python
caracteristicas =['FoodCourt','Age','RoomService','ShoppingMall','VRDeck']
```

```python
x = df[caracteristicas]

y= df['Transported']
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df['Age'])
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_val, y_train,y_val = train_test_split(x, y, test_size=0.1)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

modelo = KNeighborsClassifier(n_neighbors=2)
```

```python
scaler =StandardScaler()
X_train=scaler.fit_transform(X_train)
modelo.fit(X_train,y_train)
```

```python
predict=modelo.predict(X_val)
accuracy = accuracy_score(y_val,predict)
print(accuracy)
```
